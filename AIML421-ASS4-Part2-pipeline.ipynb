```python
data=adult_data
data.columns=['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','salary']
data.dropna(axis=0, inplace=True)  
data.head()
```

```python
import pandas as pd
adult_test =pd.read_csv("/Users/Jessie/Documents/JupyterNotebook/ass4data/part2/adult.test",header=None,skiprows=1,keep_default_na=False,na_values=['?',' ?'])
```

```python
test_data=adult_test
test_data.columns=['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','salary']
test_data.head()
```

```python
test_data=test_data.drop(['F4'],axis=1)
```

```python
test_data["salary"]=test_data["salary"].str.replace(".","")
test_data["salary"].head()
```

```python
y_test=test_data['salary']
X_test=test_data.drop(['salary'], axis=1)
X_test.shape
```

```python
y_test=y_test.str.replace(" ","")
y_test=salary_encoder.transform(y_test)
y_test
```

```python
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.datasets import fetch_openml
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_curve,auc
```

```python
#code reference: https://scikit-learn.org/stable/auto_examples/compose/plot_column_transformer_mixed_types.html

numeric_features = ["F1", "F3","F5","F11","F12","F13"]
numeric_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="median")), ("scaler", StandardScaler())]
)

categorical_features = ["F2", "F6","F7","F8","F9","F10","F14"]
categorical_transformer = OneHotEncoder(handle_unknown="ignore")

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)
```

```python
#1 KNN
from sklearn.neighbors import KNeighborsClassifier
clf = Pipeline(
    steps=[("preprocessor", preprocessor), ("classifier", KNeighborsClassifier())]
)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
y_pred_prob = clf.predict_proba(X_test)

acc=accuracy_score(y_pred, y_test)
precision=precision_score(y_pred, y_test)
recall=recall_score(y_pred, y_test)
f1=f1_score(y_pred, y_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1], pos_label=1)
roc_auc = auc(fpr, tpr)

knn_metrics=['KNN',round(acc,2),round(precision,2),round(recall,2),round(f1,2),round(roc_auc,2)]
print(knn_metrics)
```

```python
#2 GaussianNB
from sklearn.naive_bayes import GaussianNB
#!pip install mlxtend  
from mlxtend.preprocessing import DenseTransformer
clf = Pipeline(
    steps=[("preprocessor", preprocessor),('to_dense', DenseTransformer()),  ("classifier", GaussianNB())]
)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
y_pred_prob = clf.predict_proba(X_test)

acc=accuracy_score(y_pred, y_test)
precision=precision_score(y_pred, y_test)
recall=recall_score(y_pred, y_test)
f1=f1_score(y_pred, y_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1], pos_label=1)
roc_auc = auc(fpr, tpr)

nb_metrics=['Naive Bayes',acc,precision,recall,f1,roc_auc]
print(nb_metrics)
```

```python
#3 svm
from sklearn.svm import SVC
clf = Pipeline(
    steps=[("preprocessor", preprocessor), ("classifier", SVC(probability=True))]
)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
y_pred_prob = clf.predict_proba(X_test)

acc=accuracy_score(y_pred, y_test)
precision=precision_score(y_pred, y_test)
recall=recall_score(y_pred, y_test)
f1=f1_score(y_pred, y_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob[:,1], pos_label=1)
roc_auc = auc(fpr, tpr)

svm_metrics=['SVM',acc,precision,recall,f1,roc_auc]
print(svm_metrics)
```
